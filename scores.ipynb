## Loading Data

Importing libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl
```

This CSV comes from the Louisville Metro Open Data website. It is info on food establishments that have passing health department scores:

```python
passing_df = pd.read_csv('Louisville_Metro_KY_-_Restaurant_Inspection_Scores.csv')
passing_df
```

This is a separate CSV from the Louisville Metro Open Data website. This is a list of only the businesses that had a failing health department score:

```python
failed_df = pd.read_csv('Louisville_Metro_KY_-_Inspection_Violations_of_Failed_Restaurants.csv')
failed_df
```

## Cleaning Data

Dropping uneeded and redundant columns from passing_df:

```python
passing_df.drop(['PlaceName', 'Address2', 'Intersection', 'ObjectId', 'NameSearch'], axis='columns', inplace=True)
passing_df
```

Narrowing results of passing_df to only show locations such as restaurants and not schools or stores:

```python
passing_food = passing_df.query("TypeDescription == 'FOOD SERVICE'")
passing_food
```

Dropping uneeded columns from failed_df:

```python
failed_df.drop(['rpt_area_id','ObjectId'], axis='columns', inplace=True)
failed_df
```

Narrowing failed_df results to show only restaurants:

```python
failed_df.query("EstTypeDesc == 'FOOD SERVICE'", inplace=True)
failed_df
```

I noticed some duplicate data on the failed_df, so used drop_duplicates to only keep the most recent:

```python
failed_food = failed_df.drop_duplicates(subset="InspectionID", keep='last')
failed_food
```

## Merging Data

```python
merged = pd.merge(failed_food, passing_food,on='EstablishmentID',how='left')
merged.drop(['premise_city', 'premise_state', 'premise_zip', 'City', 'State', 'Zip', 'EstablishmentName'], axis =1)
merged
```

## Graphing Data

Exporting my merged datataframe via openpyxl and writer:

```python
writer = pd.ExcelWriter('merged_scores.xlsx')
merged.to_excel(writer)
writer.save()
print
```
