## Summary:  
This is a project for Code Louisville Data Analysis 2. The data being used comes from the Louisville Metro Open Data website (https://data.louisvilleky.gov). I am using this data to analyze passing restaurant health department scores vs failing restaurant inspection scores in the Louisville Kentucky Metro area.

Hypothesis: My hypothesis is that there are more passing restaurant scores than failing scores. And that there will be less failing scores in the downtown Louisville zip code as there more fine dining restaurants in that area.

## Loading Data

Importing libraries:

```python
import pandas as pd
import numpy as np
import openpyxl
```

This CSV comes from the Louisville Metro Open Data website. It is info on food establishments that have passing health department scores from the past year. This data only goes back to March 26, 2022:

```python
passing_df = pd.read_csv('Louisville_Metro_KY_-_Restaurant_Inspection_Scores.csv')
passing_df.dropna(subset=['Grade'], inplace=True)
passing_df
```

This is a separate CSV from the Louisville Metro Open Data website. This is a list of only the businesses that had a failing health department score. The available data only goes up to October 20, 2022:

```python
failed_df = pd.read_csv('Louisville_Metro_KY_-_Inspection_Violations_of_Failed_Restaurants.csv')
failed_df
```

Due to the restrictions on this data, I downloaded CSV data for March 26, 2022 through October 20, 2022. This is the timeframe of inspections that I will be comparing.

## Cleaning Data

Dropping uneeded and redundant columns from passing_df:

```python
passing_df.drop(['PlaceName', 'Address2', 'Intersection', 'ObjectId', 'NameSearch'], axis='columns', inplace=True)
passing_df
```

Narrowing results of passing_df to only show locations such as restaurants and not schools or stores:

```python
passing_df.query("TypeDescription == 'FOOD SERVICE'", inplace=True)
passing_df
```

Some restaurants have received multiple passing scores, but I only want to focus on the most recent inspections. This shows that there are 2,203 establishments coded as Food Service in this passing scores dataset:

```python
passing_food = passing_df.drop_duplicates(subset="EstablishmentID", keep='last')
passing_food
```

Dropping uneeded columns from failed_df:

```python
failed_df.drop(['rpt_area_id','ObjectId'], axis='columns', inplace=True)
failed_df
```

Narrowing failed_df results to show only restaurants:

```python
failed_df.query("EstTypeDesc == 'FOOD SERVICE'", inplace=True)
failed_df
```

Each failure type resulted in a row. For example, one business inspection may be 4 rows of data. These rows restated the score multiple times, throwing off my data. To clean this, I am dropping all duplicates. The result shows that there are 132 establishments coded as Food Service in this failed dataset:

```python
failed_food = failed_df.drop_duplicates(subset="InspectionID", keep='last')
failed_food
```

## Merging Data

Merging data on EstablishmentID column:

```python
merged = pd.merge(failed_food, passing_food,on='EstablishmentID', how='left')
merged.drop(['premise_city', 'premise_state', 'City', 'State'], axis =1, inplace=True)
merged
```

All of the failed restaurants had at least 1 passing score. We now have a data frame with 132 restaurants that at some point have both received a failed score and a passing score during this time frame. Score_x are the failing scores while score_y are the passing scores.

## Graphing Data

Exporting my merged datataframe via openpyxl and writer. I will then be reading this xlsx file into Tableau to make 3 graphs:

```python
writer = pd.ExcelWriter('merged_scores_data.xlsx')
merged.to_excel(writer)
writer.save()
print
```
